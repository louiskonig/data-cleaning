```python
import numpy as np
import pandas as pd

from IPython.display import display
```

```python
df = pd.read_csv("WELLCOME_APCspend2013_forThinkful.csv", encoding='latin1')
```

```python
def cleanup_journal_title(title):
    word_mapping = {
        'j': 'journal of',
        'med': 'medical',
        'chem': 'chemistry',
        'mol': 'molecular',
        'acs': ''
    }
    gold_mapping = {'J Med Chem': 'Journal of Medicial Chemistry'}
    
    if title in gold_mapping:
        return gold_mapping[title]
    
    tokens = str(title).lower().split()
    tokens = [word_mapping.get(tk, tk) for tk in tokens]
    tokens = [tk.capitalize() for tk in tokens if tk]
    return ' '.join(tokens)

display(len(df['Journal title'].unique()))



cleaned_titles = df['Journal title'].apply(cleanup_journal_title)
display(len(cleaned_titles.unique()))
cleaned_titles
```

```python
df_cleaned_titles = pd.DataFrame(cleaned_titles)
#cleaned_titles

grouped = df_cleaned_titles.groupby("Journal title")["Journal title"].agg([np.size]).sort_values(['size'], ascending=[False]).reset_index()
#grouped
new_list = grouped['Journal title']
grouped
```

```python
def easy_cleanup(title):
    splits = title.split()
    print('splits: {}'.format(splits))
    
    words = [[letter for letter in word if letter.isalpha()] for word in splits]
    print('words: {}'.format(words))
    
    rejoined = [''.join(letters).capitalize() for letters in words if len(letters) > 0]
    print('rejoined: {}'.format(rejoined))
    
    return ' '.join(rejoined)
```

```python
print(easy_cleanup(' hello world'))
print(easy_cleanup('junk1text_is ba@@d'))
print(easy_cleanup('123 hello'))
print(easy_cleanup('123 456 789 yo'))
```

```python
import editdistance

#titles = ['louis', 'louie', 'brad','luis']
titles = new_list
title_1s = []
title_2s = []
similar_titles = []

# NOTE: not zip. I was wrong about that, you don't just want to pair the titles together, you want every combo of titles
# The `enumerate` and [i+1:] just make sure that you don't compare the same title to itself or multiple times;
# enumerate returns the index of the title alongside the title, and the [i+1:] gets all titles after that one in the list
for i, title_1 in enumerate(titles):
    for title_2 in titles[i+1:]:
        dist = editdistance.eval(title_1, title_2)
        if 1 <= dist <= 2:
            #print("'{}' and '{}' seem very close... maybe just a typo?".format(title_1, title_2))
            title_1s.append(title_1)
            title_2s.append(title_2)

similar_titles = list(zip(title_1s, title_2s))
        
```

```python


myset = set(similar_titles)
unique_titles = list(myset)
unique_df = pd.DataFrame(unique_titles)
unique_df.to_csv('test.csv')
```
